```python
import sys
import numpy as np
import binascii as ba

# rd, rs , rt
func_dic_a = {
    "100000":"add",
    "100001":"addu",
    "100100":"and",
    "100111":"nor",
    "100101":"or",
    "101010":"slt",
    "101011":"sltu",
    "100010":"sub",
    "100011":"subu",
    "100110":"xor"
}
# rs, rt
func_dic_b = {
    "011010":"div",
    "011011":"divu",
    "011000":"mult",
    "011001":"multu"
}

# rd rt sa
func_dic_c = {
    "000000":"sll",
    "000011":"sra",
    "000010":"srl"
}

# rd rt rs
func_dic_d = {
    "000100":"sllv",
    "000111":"srav",
    "000110":"srlv"
}

# rt rs imm
func_dic_e = {
    "001000":"addi",
    "001001":"addiu",
    "001100":"andi",
    "001101":"ori",
    "001010":"slti",
    "001011":"sltiu",
    "001110":"xori"
}
# rt imm(rs)
func_dic_f = {
    "100000":"lb",
    "100100":"lbu",
    "100001":"lh",
    "100101":"lhu",
    "100011":"lw",
    "101000":"sb",
    "101001":"sh",
    "101011":"sw"
}
class InstructionI:
    
    def __init__(self, rs=0, rt=0, imm=0, func = 0):
        self.rs = rs
        self.rt = rt
        self.imm = imm
        self.func = func
    def __str__(self):
       
        if self.func in func_dic_e.values():
            strr = self.func +" $"+str(self.rt)+", $"+str(self.rs)+", "+str(self.imm)
            return strr
        elif self.func in func_dic_f.values():
            strr = self.func +" $"+str(self.rt)+", "+str(self.imm)+"($"+str(self.rs)+")"
            return strr
        elif self.func =="beq" or self.func == "bne":
            strr = self.func +" $"+str(self.rs)+", $"+str(self.rt)+", "+str(self.imm)
            return strr
        elif self.func == "lui":
            strr = self.func+ " $"+str(self.rt)+", "+str(self.imm)
            return strr
        #미완성
        elif self.func == "jal"+""
        
        elif self.func == "j":
            strr = self.func+ ""
        else:
            return self.func
    @classmethod
    def from_string(cls, str):
        func = str[:6]
        rs = int(str[6:11],2)
        rt = int(str[11:16],2)
        imm = int(str[16:],2)
        #펑션 판정
        if func in func_dic_e.keys():
            func = func_dic_e[func]
        elif func in func_dic_f.keys():
            func = func_dic_f[func]
        elif func == "000100":
            func = "beq"
        elif func == "000101":
            func = "bne"
        elif func == "001111":
            func = "lui"
        elif func == "000010":
            func = "j"
            imm = int(str[6:],2)
        elif func == "000011":
            fun = "jal"
            imm = int(str[6:],2)
        else:
            func = "unknown instruction"
            rs = ""
            rt = ""
            imm = ""
        return cls(rs, rt, imm, func)    
        
class InstructionR:
    # rd, rs , rt, sa 순으로 출력하되, 필요없는것은 출력안함
    def __init__(self, rs=0, rt=0, rd=0, sa=0, func=0):
        self.rs = str(rs)
        self.rt = str(rt)
        self.rd = str(rd)
        self.sa = str(sa)
        self.func = func
    
    
    def __str__(self):
        if self.func in func_dic_d.values():
            strr = self.func
            if self.rd != "":
                strr = strr + " $"+self.rd
            if self.rt != "":
                strr = strr + " $"+self.rt
            if self.rs != "":
                strr = strr + " $"+self.rs
            return strr
        else:
            strr = self.func
            if self.rd != "":
                strr = strr + " $"+self.rd
            if self.rs != "":
                strr = strr + " $"+self.rs
            if self.rt != "":
                strr = strr + " $"+self.rt
            if self.sa != "":
                strr = strr + " "+self.sa
            return strr
        
    
    @classmethod
    def from_string(cls, str):
        rs = int(str[6:11],2)
        rt = int(str[11:16],2)
        rd = int(str[16:21],2)
        sa = int(str[21:26],2)
        func = str[26:]
        # 펑션 판정 
        if func in func_dic_a.keys():
            func = func_dic_a[func]
            sa = ""
        elif func in func_dic_b.keys():
            func = func_dic_b[func]
            sa = ""
            rd = ""
        elif func in func_dic_c.keys():
            func = func_dic_c[func]
            rs = ""
        elif func in func_dic_d.keys():
            func = func_dic_d[func]
            sa = ""
        elif func == "001100":
            func = "syscall"
            rs = ""
            rt = ""
            rd = ""
            sa = ""
        elif func == "001001":
            func = "jalr"
            rt = ""
            sa = ""
        elif func == "001000":
            func = "jr"
            rd = ""
            rt = ""
            sa = ""
        elif func == "010000":
            func = "mfhi"
            rs = ""
            rt = ""
           
            sa = ""
        elif func == "001010":
            func = "mflo"
            rs = ""
            rt = ""
            rd = ""
            sa = ""
        elif func == "010001":
            func = "mthi"
           
            rt = ""
            rd = ""
            sa = ""
        elif func == "010011":
            func = "mtlo"
            
            rt = ""
            rd = ""
            sa = ""
       
        else:
            func = "unknown instruction"
            rs = ""
            rt = ""
            rd = ""
            sa = ""
        
        inst = cls(rs, rt, rd, sa, func)
        return inst
    
    
def hex2bin(str):
    bin_str = bin((int(str,16)))
    bin_str = bin_str[2:]
    num = 32 - len(bin_str)
    zero = "0"*num
    return zero+bin_str


# 이진 파일을 16진수 바이트 타입으로 컨버팅해줘 받아옴
with open("test1.bin", "rb") as binary:
    test=(ba.b2a_hex(binary.read()))


    
instructions = []
```
